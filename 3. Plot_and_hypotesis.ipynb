### Add Count of number of badges to the previews dataframe

```python
# Adding count of number of badges per school to badges dataframe
df_badges_count = badges.groupby('school_id')['name'].nunique().reset_index()
df_badges_count
```

## Spearman Correlation method to check some correlations hypotheses

```python
comments_numeric_corr = comments_avgprices.corr(method='spearman')
comments_numeric_corr.style.background_gradient(cmap='RdYlGn').format('{:,.2f}')
```

## Hypothesis Testing about corelation between Job support and Overall score

```python
# Hypothesis testing for correlation between job support and overall scores
correlation, p_value = stats.pearsonr(comments_numeric['jobSupport'], comments_numeric['overallScore'])

alpha = 0.05

# Check if the p-value is less than the significance level
if p_value < alpha:
    print(f"There is a significant correlation between job support and overall scores (p-value: {p_value:.4f})")
else:
    print(f"No significant correlation between job support and overall scores (p-value: {p_value:.4f})")
```
